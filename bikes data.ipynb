```python
p_year=(pd.unique(bikes['model_year']))
for i in range(0, len(p_year)):    
    for j in range(i+1, len(p_year)):    
        if(p_year[i] > p_year[j]):    
            temp = p_year[i]   
            p_year[i] = p_year[j]    
            p_year[j] = temp   
     
     
#Displaying elements of the array after sorting    
    
print("Elements of array sorted in ascending order: ");   
for i in range(0, len(p_year)):    
    print(p_year[i], end=" ")
```
