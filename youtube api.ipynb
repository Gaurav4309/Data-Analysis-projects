```python
#individual video data for 1st video
video_id=response['items'][0]['id']['videoId']
video_title=response['items'][0]['snippet']['title']
video_date=response['items'][0]['snippet']['publishedAt']
video_date=video_date.split('T')[0]

#individual video metrics
stats_url='https://www.googleapis.com/youtube/v3/videos?key='+api_key+'&part=statistics&id='+video_id
stats_response=requests.get(stats_url).json()
view_count=stats_response['items'][0]['statistics']['viewCount']
like_count=stats_response['items'][0]['statistics']['likeCount']
comment_count=stats_response['items'][0]['statistics']['commentCount']



```

```python
#individual video data for all videos on page 1
for video in response['items']:
    if video['id']['kind']=='youtube#video':
    
        video_id=video['id']['videoId']
        video_title=video['snippet']['title']
        video_date=video['snippet']['publishedAt']
        video_date=video_date.split('T')[0]

        #individual video metrics
        stats_url='https://www.googleapis.com/youtube/v3/videos?key='+api_key+'&part=statistics&id='+video_id
        stats_response=requests.get(stats_url).json()
        view_count=stats_response['items'][0]['statistics']['viewCount']
        like_count=stats_response['items'][0]['statistics']['likeCount']
        comment_count=stats_response['items'][0]['statistics']['commentCount']
        
        #adding data to dataframe df
        df=df.append({'video_id':video_id,'video_title':video_title,'video_date':video_date,'view_count':view_count,
                     'like_count':like_count,'comment_count':comment_count},ignore_index=True)

```
